# 请求+获取网页数据	

```python
import requests   # 导入requests包

url = 'https://movie.douban.com/top250'
# 尝试加入user-agent
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
            AppleWebKit/537.36 (KHTML, like Gecko) \
           Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0'}
resp = requests.get(url, headers=headers)  # 设置参数headers
resp #返回码200说明访问成功
```

```python
# 查看返回内容，搜索该页面的“肖申克的救赎”是否在返回的文本中，确认与网页一致
resp.text
```

```python
# 将返回内容放入html中，html为str
html = resp.text
html
type(html)
```

# 解析数据

```python
from pyquery import PyQuery as pq #导入pyquery包
```

```python
doc = pq(html) # 将html字符串转换为pyquery数据，便于解析
type(doc)   # 数据类型为pyquery.pyquery.PyQuery
doc.text()    # 数据变“干净”了
```

**目标信息源代码（部分）**

```python
titles = doc('div.hd a') # 找到所有书名
print(titles)
```

```python
for title in titles:
    print(pq(title).text())
```
