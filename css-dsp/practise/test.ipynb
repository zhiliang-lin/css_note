```python
import requests
from bs4 import BeautifulSoup

# 定义请求头
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'
}

# 发送请求
response = requests.get("https://book.douban.com/top250", headers=headers)

# 使用 BeautifulSoup 解析 HTML
soup = BeautifulSoup(response.text, 'html.parser')

# 提取数据
books = soup.select("td:nth-of-type(2)")

for book in books:
    title = book.select_one("a").text.strip()
    info = book.select_one("p.pl").text.strip()
    rating = book.select_one("span.rating_nums").text.strip()
    count = book.select_one("span.pl").text.strip().split()[0]
    quote = book.select_one("p.quote").text.strip() if book.select_one("p.quote") else ""
    
    print("Title:", title)
    print("Info:", info)
    print("Rating:", rating)
    print("Count:", count)
    print("Quote:", quote)
    print("-" * 50)
```
