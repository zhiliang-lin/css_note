```python
import requests
from pyquery import PyQuery as pq
import csv
```

# 请求数据的函数

```python
def get_html(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    r = requests.get(url,headers=headers)
    html = r.text
    return html
```

# 构建url_list的函数

```python
def generate_url_list(page_num): # 要抓多少页

    url_list = []
    url_template = "https://since1989.org/page/{page}"


    for p in range(1, page_num+1):
        url = url_template.format(page=p)
        url_list.append(url)
    
    return url_list
    
```

# 解析数据的函数

```python
def extract_article_list(html):

    doc = pq(html)
    articles = doc('article')

    article_list =[]

    for article in articles.items():
        title = article("h2").text()
        time = article("time.entry-date").text()
        summary = article("div.entry-content p").text()
        tag = article("span.cat-links").text()
        link = doc('h2.entry-title a').attr('href')

    
        artcle_dict = {"title": title, 
                       "time": time,
                       "summary": summary, 
                       "tag": tag, 
                       "link": link}

        article_list.append(artcle_dict)
        
    return article_list
    
```

```python
# 函数：爬虫主函数  main(filename)
# 参数说明：filename为文件名称
# 仅执行命令，不返回任何值
def main(filename):         # 添加：定义函数名
    print('开始采集博客我是王掌柜！')              # 添加：说明“开始采集豆瓣读书top 250书籍！”
    
    # 生成特定数量页码的网址url_list
    url_list = generate_url_list(796)  #抓前796页
    
    # 打开文件
    file = open(filename, 'a+', encoding='utf-8', newline='')     # 修改：将文件名称，设定为函数的参数
    fieldnames = ['title','summary','time','tag','link']  
    writer = csv.DictWriter(file, fieldnames=fieldnames)    
    writer.writeheader() 

    # 对所有网址url_list循环步骤2-4
    for url in url_list:
        print('正在采集：{url}'.format(url=url))    # 添加：说明“正在采集<url>”
        html = get_html(url)
        article_list = extract_article_list(html)
        for article in article_list:
            writer.writerow(article)

    file.close()

    print('采集完毕！')            # 添加：说明“采集完毕！”
```

```python
main("since1989_p1-p796.csv")
```
