```python
from selenium import webdriver
from pyquery import PyQuery as pq

```

```python
# 测试
browser = webdriver.Firefox()
# browser.get('https://www.baidu.com/')
```

# 案例1：在百度新闻（news.baidu.com）搜索“小米”，获取首页新闻标题


```python

url = 'https://www.baidu.com/s?rtt=1&bsst=1&cl=2&tn=news&rsv_dl=ns_pc&word=%E5%B0%8F%E7%B1%B3&x_bfe_rqs=032080000000000000000000000000000000000000000008&x_bfe_tjscore=0.080000&tngroupname=organic_news&newVideo=12&goods_entry_switch=1&pn=10'


browser.get(url)
html = browser.page_source
print(html)
```

```python
doc = pq(html)

titles = doc('h3')
for title in titles.items():
    print(title.text())
```

```python
import csv

# 创建一个 CSV 文件并打开以写入模式
with open('baidu_news_titles.csv', 'w', newline='', encoding='utf-8') as csvfile:
    # 定义 CSV 写入器
    csv_writer = csv.writer(csvfile)
    
    # 遍历 titles 字典中的每个标题，并逐行写入 CSV 文件
    for title in titles.items():
        csv_writer.writerow([title.text()])
```

# 练习1：知乎热搜榜标题

```python
#先手动登录知乎

url = 'https://www.zhihu.com/hot'

browser.get(url)
html = browser.page_source
print(html)
```

```python
doc = pq(html)
titles = doc('h2.HotItem-title')

for title in titles.items():
    print(title.text())
```

```python
import csv

# 创建一个 CSV 文件并打开以写入模式
with open('zhihu_hot.csv', 'w', newline='', encoding='utf-8') as csvfile:
    # 定义 CSV 写入器
    csv_writer = csv.writer(csvfile)
    
    # 遍历 titles 字典中的每个标题，并逐行写入 CSV 文件
    for title in titles.items():
        csv_writer.writerow([title.text()])
```
