```python
import numpy as np
import matplotlib.pyplot as plt

# 生成数据
x = np.linspace(0.01, 5, 100)  # 避免log(0)的情况，从一个小的正数开始
y = np.log(x)

# 绘制log函数的图形
plt.plot(x, y, label='log(x)')

# 添加标签和标题
plt.xlabel('x')
plt.ylabel('log(x)')
plt.title('Logarithmic Function')


# 显示图例
plt.legend()

# 显示图形
plt.show()
```
