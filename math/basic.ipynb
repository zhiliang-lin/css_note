```python
import numpy as np
import matplotlib.pyplot as plt

# 生成数据
x = np.linspace(0.01, 5, 100)  # 避免log(0)的情况，从一个小的正数开始
y = np.log(x)

# 绘制log函数的图形
plt.plot(x, y, label='log(x)')

# 添加标签和标题
plt.xlabel('x')
plt.ylabel('log(x)')
plt.title('Logarithmic Function')


# 显示图例
plt.legend()

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 定义幂函数
def power_function(x, n):
    return x**n

# 生成 x 值
x_values = np.linspace(-2, 2, 100)  # 生成在区间[-2, 2]上的100个点

# 绘制不同幂次的幂函数曲线
plt.figure(figsize=(8, 6))

for n in range(1, 6):
    y_values = power_function(x_values, n)
    plt.plot(x_values, y_values, label=f'$y=x^{n}$')

    # 在曲线附近添加数学公式
    text_x = x_values[np.argmax(y_values)]  # 选择曲线上 y 值最大的点
    text_y = np.max(y_values)
    plt.text(text_x, text_y, f'$y=x^{n}$', fontsize=10, ha='left', va='bottom')

# 添加标签和标题
plt.xlabel('x')
plt.ylabel('y')
plt.title('Power Functions')
plt.legend()

# 显示图形
plt.grid(True)
plt.show()
```
