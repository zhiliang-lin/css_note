
# 🐍 Python Environment Management for Jupyter

Managing environments allows you to isolate project dependencies so that libraries from one project don't conflict with another.

---

## 1. Why use environments?
*   **Isolation:** Project A needs `pandas 1.0`, Project B needs `pandas 2.0`.
*   **Reproducibility:** Ensures your code runs on other machines.
*   **Safety:** Prevents breaking your system-level Python installation.

---

## 2. Option A: Using `venv` (Standard Python)
*Best for lightweight, standard Python projects.*

#### 1. Create the environment
Open your terminal (or Command Prompt) and run:
```bash
# syntax: python -m venv <name_of_env>
python -m venv my_project_env
```

#### 2. Activate the environment
*   **Windows:** `my_project_env\Scripts\activate`
*   **Mac/Linux:** `source my_project_env/bin/activate`

*(You will see the environment name in parenthesis in your terminal prompt).*

#### 3. Install your libraries
```bash
pip install pandas numpy scikit-learn
```

---

## 3. Option B: Using Conda (Anaconda/Miniconda)
*Best for Data Science, Heavy Machine Learning, and non-Python dependencies.*

#### 1. Create the environment
```bash
# syntax: conda create -n <name> python=<version>
conda create -n my_conda_env python=3.10
```

#### 2. Activate the environment
```bash
conda activate my_conda_env
```

#### 3. Install libraries
```bash
conda install pandas numpy scikit-learn
```

---

## 4. ⭐️ CRITICAL: Connecting to Jupyter
By default, Jupyter doesn't know your new environment exists. You need to create a **Kernel**.

#### Step 1: Install ipykernel (inside your new env)
Make sure your environment is **activated**, then run:

```bash
pip install ipykernel
```

#### Step 2: Add the kernel to Jupyter
This command makes your environment visible in the Jupyter "Kernel" dropdown menu.

```bash
# syntax: python -m ipykernel install --user --name=<env_name> --display-name "<Visible Name>"

python -m ipykernel install --user --name=my_project_env --display-name "Python (My Project)"
```

---

## 5. Usage in Jupyter Notebook
1.  Launch Jupyter Notebook: `jupyter notebook`
2.  Open your notebook.
3.  Go to the Menu: **Kernel** $\rightarrow$ **Change Kernel** $\rightarrow$ Select **Python (My Project)**.

---

## 6. Useful Commands Cheat Sheet

| Action | `venv` Command | `conda` Command |
| :--- | :--- | :--- |
| **List Envs** | (View folder structure) | `conda env list` |
| **Deactivate** | `deactivate` | `conda deactivate` |
| **Export reqs** | `pip freeze > requirements.txt` | `conda env export > environment.yml` |
| **Install reqs** | `pip install -r requirements.txt` | `conda env create -f environment.yml` |
| **Remove Kernel**| `jupyter kernelspec uninstall my_env_name` | `jupyter kernelspec uninstall my_env_name` |

*(也可以通过 `pip install uv` 安装，但独立安装更推荐)*

#### 2. 虚拟环境管理 (`uv venv`)
`uv` 能够瞬间创建虚拟环境。

* **创建默认环境**：
    ```bash
    uv venv
    ```
    *会在当前目录生成一个 `.venv` 文件夹。*
* **指定 Python 版本创建**：
    ```bash
    uv venv --python 3.11
    ```
* **激活环境**（传统方式）：
    ```cmd
    .venv\Scripts\activate
    ```
    *(注：使用 `uv run` 和 `uv pip` 时通常**不需要**手动激活)*

#### 3. 极速包管理 (`uv pip`)
只要当前目录下有 `.venv` 文件夹，`uv pip` 就会自动将包安装到该虚拟环境中，无需激活。

* **安装包**：
    ```bash
    uv pip install requests pandas
    ```
* **从 requirements.txt 安装**：
    ```bash
    uv pip install -r requirements.txt
    ```
* **导出当前环境的依赖**：
    ```bash
    uv pip freeze > requirements.txt
    ```
* **卸载包**：
    ```bash
    uv pip uninstall requests
    ```

#### 4. 执行命令与脚本 (`uv run`)
`uv run` 会自动寻找并使用当前目录下的 `.venv` 环境来执行命令，非常适合写在批处理脚本里或日常快速调用。

* **运行 Python 脚本**：
    ```bash
    uv run main.py
    ```
* **运行环境中的工具**（例如 jupyter、pytest、fastapi）：
    ```bash
    uv run jupyter lab
    ```

#### 5. 现代项目管理 (类似 npm 或 Poetry)
如果你想对项目进行更规范的管理（使用 `pyproject.toml`），`uv` 也提供了开箱即用的支持。

* **初始化新项目**：
    ```bash
    uv init
    ```
    *会生成 `pyproject.toml`、`hello.py` 等基础文件。*
* **添加依赖**（自动安装并写入 `pyproject.toml`）：
    ```bash
    uv add requests
    ```
* **添加开发依赖**（仅开发环境需要，如 pytest）：
    ```bash
    uv add --dev pytest
    ```
* **移除依赖**：
    ```bash
    uv remove requests
    ```
* **同步整个项目的依赖**（确保当前虚拟环境和 `pyproject.toml` 完全一致）：
    ```bash
    uv sync
    ```

#### 💡 核心逻辑总结
1. **建环境**：`uv venv`
2. **装依赖**：`uv pip install <包名>` (临时脚本) 或 `uv add <包名>` (正规项目)
3. **跑代码**：`uv run <命令或脚本>`
