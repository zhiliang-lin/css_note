```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 中文字体设置-黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# 示例 1: 基本折线图
# 生成 X 和 Y 轴的数据
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 创建折线图
plt.plot(x, y)
# 添加标题和轴标签
plt.title('基本折线图')
plt.xlabel('X 轴')
plt.ylabel('Y 轴')

# 显示图表
plt.show()
```

```python

# 示例 2: 散点图
# 生成 X 和 Y 轴的随机数据
x = np.random.rand(50)
y = np.random.rand(50)

# 为每个点生成随机颜色和大小
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

# 创建散点图
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)

# 添加标题和轴标签
plt.title('散点图')
plt.xlabel('X 轴')
plt.ylabel('Y 轴')

# 显示图表
plt.show()
```

```python

# 示例 3: 条形图
# 定义类别及其对应的值
categories = ['类别 A', '类别 B', '类别 C']
values = [3, 7, 5]

# 创建条形图
plt.bar(categories, values, color='skyblue')

# 添加标题和轴标签
plt.title('条形图')
plt.xlabel('类别')
plt.ylabel('值')

# 显示图表
plt.show()
```
