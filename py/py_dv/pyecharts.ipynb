```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Pie, Boxplot, Bar3D, Scatter
import numpy as np
```

# 柱状图

```python

# 示例数据
categories = ['A', 'B', 'C', 'D', 'E']
data = [10, 20, 15, 25, 30]

# 创建柱状图
bar_chart = (
    Bar()
    .add_xaxis(categories)
    .add_yaxis("柱状图示例", data)
    .set_global_opts(title_opts=opts.TitleOpts(title="My Bar Chart"))
)

# 生成 HTML 文件（可选）
bar_chart.render("bar_chart.html")

# 在 Jupyter Notebook 中显示图表（可选）
bar_chart.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.render import notebook
import numpy as np

# Example data
data = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data, bins=30)

# Convert edges to strings for y-axis labels
y_labels = [f"{edges[i]:.2f}-{edges[i+1]:.2f}" for i in range(len(edges)-1)]

# Create Horizontal Bar chart
bar_chart = (
    Bar()
    .add_xaxis(hist.tolist())
    .add_yaxis("Histogram", y_labels, color="skyblue", category_gap="80%")  # Adjust gap for better visualization
    .reversal_axis()  # Reverse the axis to make it horizontal
    .set_global_opts(title_opts=opts.TitleOpts(title="Histogram"))
)

# Render the chart in the Jupyter Notebook
notebook.render_notebook(bar_chart)
```

# 饼状图

```python

# Pie chart
labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]
pie = (
    Pie()
    .add("", [list(z) for z in zip(labels, sizes)])
    .set_global_opts(title_opts=opts.TitleOpts(title="Pie Chart"))
)
pie.render_notebook()
```

# 箱型图

```python

# Boxplot
boxplot = (
    Boxplot()
    .add_xaxis(["Boxplot"])
    .add_yaxis("", [data.tolist()])
    .set_global_opts(title_opts=opts.TitleOpts(title="Boxplot"))
)
boxplot.render_notebook()
```

```python
# Stacked bar chart
categories = ['Category A', 'Category B', 'Category C']
values1 = [20, 30, 25]
values2 = [15, 25, 30]

bar3d = (
    Bar3D()
    .add("", categories, ["Group 1"] * len(categories), values1)
    .add("", categories, ["Group 2"] * len(categories), values2)
    .set_global_opts(title_opts=opts.TitleOpts(title="Stacked Bar Chart"))
)
bar3d.render_notebook()
```

```python
# Scatter plot
x = np.random.randn(100)
y = 10 * x + np.random.randn(100)

scatter = (
    Scatter()
    .add_xaxis(x.tolist())
    .add_yaxis("Scatter Plot", y.tolist(), symbol_size=8)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Scatter Plot"),
        xaxis_opts=opts.AxisOpts(min_=min(x), max_=max(x)),  # Adjust x-axis scale
        yaxis_opts=opts.AxisOpts(min_=min(y), max_=max(y)),  # Adjust y-axis scale
    )
)
scatter.render_notebook()
```
