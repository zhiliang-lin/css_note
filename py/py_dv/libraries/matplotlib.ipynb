# Matplotlib 的绘图功能可以分为两部分：

- 绘图类函数：如 plot()、scatter() 等，用于创建具体图表。
- 辅助类函数：如 title()、xlabel() 等，用于添加标题、标签、图例等内容，增强图表的可读性。

## 辅助类函数

-  `plt.title()` ：设置图表的标题。
- `plt.xlabel()` ：设置横轴的标签
- `plt.ylabel()` ：设置纵轴的标签。
- `plt.grid()` ：添加网格线。
- `plt.legend()` ：添加图例。
- `plt.savefig()` ：将图表保存为文件。

以下是 Matplotlib 辅助类函数的常见参数详细介绍，以及如何使用这些参数来调整图表的外观和风格。

---

### **1. `plt.title()`**
设置图表的标题。

**语法**：
```python
plt.title(label, loc='center', fontsize=None, color=None, pad=None)
```

**参数说明**：
- `label`：标题内容（字符串）。
- `loc`：标题位置，可选 `'center'`（默认）、`'left'` 或 `'right'`。
- `fontsize`：字体大小（数字）。
- `color`：字体颜色（如 `'red'`、`'#FF5733'`）。
- `pad`：标题与图表的垂直间距（单位为像素）。

**示例**：
```python
plt.title('Sales Analysis', loc='left', fontsize=16, color='darkblue', pad=10)
```

---

### **2. `plt.xlabel()` 和 `plt.ylabel()`**
设置横轴和纵轴的标签。

**语法**：
```python
plt.xlabel(label, fontsize=None, color=None, labelpad=None)
plt.ylabel(label, fontsize=None, color=None, labelpad=None)
```

**参数说明**：
- `label`：轴标签内容（字符串）。
- `fontsize`：字体大小。
- `color`：字体颜色。
- `labelpad`：标签与轴线的间距（单位为像素）。

**示例**：
```python
plt.xlabel('Time (s)', fontsize=12, color='green', labelpad=10)
plt.ylabel('Speed (m/s)', fontsize=12, color='purple', labelpad=15)
```

---

### **3. `plt.grid()`**
在图表上添加网格线。

**语法**：
```python
plt.grid(b=None, which='major', axis='both', color=None, linestyle=None, linewidth=None, alpha=None)
```

**参数说明**：
- `b`：是否显示网格线（布尔值，默认 `True`）。
- `which`：选择网格线类型，`'major'`（默认）表示主刻度线，`'minor'` 表示次刻度线。
- `axis`：网格线应用的轴，`'both'`（默认）、`'x'` 或 `'y'`。
- `color`：网格线颜色。
- `linestyle`：网格线样式（如 `'-'`、`'--'`、`':'`）。
- `linewidth`：网格线宽度。
- `alpha`：网格线透明度（0 到 1）。

**示例**：
```python
plt.grid(which='major', color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
```

---

### **4. `plt.legend()`**
在图表上添加图例。

**语法**：
```python
plt.legend(loc='best', fontsize=None, title=None, title_fontsize=None, frameon=True, shadow=False)
```

**参数说明**：
- `loc`：图例位置，可选：
  - `'best'`（默认）：自动选择最佳位置。
  - 其他选项如 `'upper right'`、`'lower left'`、`'center'` 等。
- `fontsize`：图例字体大小。
- `title`：图例标题。
- `title_fontsize`：图例标题字体大小。
- `frameon`：是否显示边框（布尔值，默认 `True`）。
- `shadow`：是否在边框下添加阴影（布尔值）。

**示例**：
```python
plt.legend(loc='upper right', fontsize=10, title='Legend', title_fontsize=12, frameon=True, shadow=True)
```

---

### **5. `plt.xticks()` 和 `plt.yticks()`**
设置刻度标签。

**语法**：
```python
plt.xticks(ticks=None, labels=None, rotation=None, fontsize=None, color=None)
plt.yticks(ticks=None, labels=None, rotation=None, fontsize=None, color=None)
```

**参数说明**：
- `ticks`：刻度位置（列表）。
- `labels`：刻度标签（列表）。
- `rotation`：刻度标签旋转角度（如 `45`、`90`）。
- `fontsize`：刻度标签字体大小。
- `color`：刻度标签颜色。

**示例**：
```python
plt.xticks(ticks=[1, 2, 3, 4], labels=['Q1', 'Q2', 'Q3', 'Q4'], rotation=45, fontsize=10, color='blue')
plt.yticks(fontsize=10, color='green')
```

---

### **6. `plt.savefig()`**
将图表保存为文件。

**语法**：
```python
plt.savefig(fname, dpi=None, bbox_inches=None, transparent=False, facecolor='w', edgecolor='w')
```

**参数说明**：
- `fname`：文件名（包括路径和扩展名，如 `'chart.png'`）。
- `dpi`：分辨率（每英寸点数，默认为 `100`）。
- `bbox_inches`：保存的图表范围，`'tight'` 会减少空白边距。
- `transparent`：是否保存为透明背景（布尔值）。
- `facecolor`：背景颜色。
- `edgecolor`：边框颜色。

**示例**：
```python
plt.savefig('my_plot.png', dpi=300, bbox_inches='tight', transparent=True)
```

---

### **7. `plt.subplots_adjust()`**
调整子图的布局。

**语法**：
```python
plt.subplots_adjust(left=None, right=None, top=None, bottom=None, wspace=None, hspace=None)
```

**参数说明**：
- `left`、`right`、`top`、`bottom`：子图与图边缘的间距（0 到 1）。
- `wspace`：子图之间的水平间距。
- `hspace`：子图之间的垂直间距。

**示例**：
```python
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.4, hspace=0.6)
```


````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 15, 20, 25, 30]
y2 = [5, 10, 15, 20, 25]

# 绘制两条折线
plt.plot(x, y1, label='Dataset 1', color='blue', linestyle='--', marker='o')
plt.plot(x, y2, label='Dataset 2', color='green', linestyle='-', marker='s')

# 添加标题和轴标签
plt.title('Sales Data', fontsize=16, color='darkred', pad=20)
plt.xlabel('Time (Months)', fontsize=12, labelpad=10)
plt.ylabel('Sales ($)', fontsize=12, labelpad=10)

# 设置网格线和图例
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend(loc='upper left', fontsize=10, title='Legend', title_fontsize=12)

# 保存并显示图表
# plt.savefig('custom_chart.png', dpi=300, bbox_inches='tight')
plt.show()
````

# 饼状图

````python
# 创建示例 DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D', 'A', 'B', 'C', 'D', 'A', 'C'],
    'Value': [50, 30, 40, 20, 60, 45, 35, 25, 55, 50]
}
df = pd.DataFrame(data)

# 绘制饼状图：按类别统计计数
category_counts = df['Category'].value_counts()

plt.figure(figsize=(6, 6))
category_counts.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'coral'], startangle=140)
plt.title('Pie Chart of Categories')
plt.ylabel('')  # 去掉默认的y轴标签
plt.show()
````

````python

# 示例 3: 条形图
# 定义类别及其对应的值
categories = ['类别 A', '类别 B', '类别 C']
values = [3, 7, 5]

# 创建条形图
plt.bar(categories, values, color='skyblue')

# 添加标题和轴标签
plt.title('条形图')
plt.xlabel('类别')
plt.ylabel('值')

# 显示图表
plt.show()
````

# 折线图

````python

# 示例 2: 散点图
# 生成 X 和 Y 轴的随机数据
x = np.random.rand(50)
y = np.random.rand(50)

# 为每个点生成随机颜色和大小
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

# 创建散点图
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)

# 添加标题和轴标签
plt.title('散点图')
plt.xlabel('X 轴')
plt.ylabel('Y 轴')

# 显示图表
plt.show()
````

# 雷达图

````python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个示例数据
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [4, 2, 5, 3, 1]

# 计算角度
theta = np.linspace(0, 2*np.pi, len(categories), endpoint=False)

# 将最后一个点闭合
values = np.concatenate((values, [values[0]]))
theta = np.concatenate((theta, [theta[0]]))

# 画雷达图
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(theta, values, color='b', alpha=0.1)

# 添加刻度标签
ax.set_xticks(theta[:-1])
ax.set_xticklabels(categories)

# 显示图形
plt.show()
````

- **多个组别**

````python
import numpy as np
import matplotlib.pyplot as plt

# 创建示例数据
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values_series1 = [4, 2, 5, 3, 1]
values_series2 = [2, 5, 1, 4, 3]

# 计算角度
theta = np.linspace(0, 2*np.pi, len(categories), endpoint=False)

# 将最后一个点闭合
values_series1 = np.concatenate((values_series1, [values_series1[0]]))
values_series2 = np.concatenate((values_series2, [values_series2[0]]))
theta = np.concatenate((theta, [theta[0]]))

# 画雷达图
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# 画第一个数据系列
ax.fill(theta, values_series1, color='b', alpha=0.1, label='Series 1')
# 画第二个数据系列
ax.fill(theta, values_series2, color='r', alpha=0.1, label='Series 2')

# 添加刻度标签
ax.set_xticks(theta[:-1])
ax.set_xticklabels(categories)

# 添加图例
ax.legend()

# 显示图形
plt.show()
````
