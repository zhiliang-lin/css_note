## 关于axis

```python
A = np.arange(10).reshape(2,5)
A
```

- **axis = 0 指的是列**

```python
# 每列求和
A.sum(axis=0)
```

- **axis = 1 指的是行**

```python
# 每行求和
A.sum(axis=1)
```
